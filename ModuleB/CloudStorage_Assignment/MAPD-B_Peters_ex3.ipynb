# Exercise 3 - Cloud Storage

## 3.1) Write a program in C, C++, Python or using ROOT, which simulates the placement of 10GB files to random locations and account the used space on each hard disk. Once the first hard disk is full, you stop to place files.

```python
#set the random-seed
rn.seed(12347789)
```

```python
#create the disks:
N_disk = 10
size_disk = 1000
disks = [0 for i in range(N_disk)]
keys = [i for i in range(0,N_disk)]
block_size = 10  #10 GB size!!

#fill the disks with 10 GB files untill one of them is full:
while (max(disks)<size_disk):
    a = rn.randint(0,N_disk-1)
    disks[a] = disks[a] + block_size


#Visualize the distribution:
dictionary = dict(zip(keys, disks))
fig, ax = plt.subplots(figsize=(12,6)) # Create a figure and an axes.
ax.bar(dictionary.keys(),dictionary.values())  
ax.set_xlabel('disk ID')  # Add an x-label to the axes.
ax.set_ylabel('Occupied size (GB)')  # Add a y-label to the axes.
ax.set_title("Occupied Gigabyte in the ten disks with blocks of size {:d} GB".format(block_size))  # Add a title to the axes.


#3.1a How many files did you manage to place?
N_files = sum(disks)/block_size
print('The number of files we manage to place is: {:d}'.format(int(N_files)))

#3.1b What is the percentage of total used space on all hard disks in the moment the first disk is full?
P_space = sum(disks)/(size_disk*N_disk)*100
print('The percentage of total used space on all hard disks is: {:.2f} %'.format(P_space))
```
